```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_excel("C:\\Users\\exam2\\Downloads\\Adidas_US_Sales_Datasets_1_.xlsx")
data
```

1.	List all the retailers with retailer id.

```python
data[['Retailer','Retailer ID']]
```

2.	List all the retailers in every region.

```python
data.groupby(['Region'])['Retailer'].unique().apply(list)
```

3.	List the retailers in every city of a state.

```python
data.groupby(['State','City'])['Retailer'].apply(list)
```

4.	List the products sold by the retailer in every city.

```python
data.groupby(['City','Retailer'])['Product'].apply(list)
```

5.	Find the total sales of every retailer.

```python
data.groupby(['Retailer'])['Total Sales'].count()
```

6.	Find the total sales of the retailers in every city along with profit.

```python
data.groupby(['Retailer','City'])['Total Sales','Operating Profit'].sum()
```

7.	Find the total sales &profit of each product sold by the retailer.

```python
data.groupby(['Retailer','Product'])['Total Sales','Operating Profit'].sum()
```

8.	Find the units sold, total sales &profit of the products sold between the dates 1/1/2020 and 4/15/2020

```python
k=data[(data['Invoice Date']<="2020-04-15") & (data['Invoice Date']>="2020-01-01")]
k.filter(['Units Sold','Total Sales','Operating Profit'])
```

9.	Find the no of units sold of each product by each retailer in every city.

```python
data.groupby(['City','Retailer','Product'])['Units Sold'].apply(list)
```

10.	Find the products with different price per unit in different cities with proper information.

```python
data.groupby(['City','Price per Unit'])['Product'].unique()
```

11.	Find the total sales &profits of all products in every month.

```python
data.groupby([pd.Grouper(key='Invoice Date',axis=0,freq='M'),'Product'])['Total Sales','Operating Profit'].sum()
```

12.	Find the total sales &profit of the products in different sales methods in each city

```python
data.groupby(['City','Sales Method'])['Total Sales','Operating Profit'].sum()
```

13.	Find the retailers who sold the same product with different prices in different cities.

```python
data.groupby(['City','Product','Price per Unit'])['Retailer'].apply(list)
```

14.	Find the products whose sales raises in every month.

```python
data.groupby(['Product',pd.Grouper(key='Invoice Date',axis=0,freq='M')])['Total Sales'].sum()
```

15.	Find the retailers whose profit increased every month.

```python
data.groupby(['Product',pd.Grouper(key='Invoice Date',axis=0,freq='M')])['Operating Profit'].sum().groupby('Product').diff()
```
